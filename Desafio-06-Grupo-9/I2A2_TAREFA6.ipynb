# **:: Atividade 2 ‚Äì Limpeza e Padroniza√ß√£o**

<h4 align="justify">A diversidade de origens e destinos pode gerar confus√µes quando informa√ß√µes semelhantes aparecem de forma inconsistente, prejudicando relat√≥rios ambientais e log√≠sticos.</h4>
<h4 align="justify">Identifique e trate valores inconsistentes (ex.: "reciclagem formal" e "Reciclagem Formal"), corrija erros de digita√ß√£o e padronize todas as colunas categ√≥ricas. Em seguida, elimine duplicatas mantendo apenas registros v√°lidos.</h4>

```python
# Instala a biblioteca unicode no Colab
!pip install unidecode

# OBS: Pandas j√° foi importado na Atividade 1
from unidecode import unidecode
import re

# Fazemos uma c√≥pia da DB para fazer a padroniza√ß√£o
df_padronizacao = df.copy()

print("üîÑ Limpeza e Padroniza√ß√£o\n")

print("1Ô∏è‚É£ Convers√£o da coluna Data_Descarte para datetime")
# Transformamos de object para datetime e depois mantemos apenas a parte da data em si
# Tratamos poss√≠veis formatos inconsistentes
df_padronizacao["Data_Descarte"] = pd.to_datetime(df_padronizacao["Data_Descarte"], errors="coerce")

# Conferir os primeiros valores para validar a convers√£o
print("‚úÖ Primeiras linhas da coluna Data_Descarte ap√≥s convers√£o:")
display(df_padronizacao[["Data_Descarte"]].head(10))

# Criar coluna formatada para visualiza√ß√£o em formato brasileiro
df_padronizacao['Data_Descarte_BR'] = df_padronizacao['Data_Descarte'].dt.strftime('%d/%m/%Y')

# Conferir primeiras linhas
display(df_padronizacao[['Data_Descarte', 'Data_Descarte_BR']].head(10))


print("\n2Ô∏è‚É£ Diagn√≥stico inicial dos valores categ√≥ricos")
colunas_categoricas = ["Tipo_Eletronico", "Origem", "Destino_Final", "Municipio", "Educacao_Ambiental"]

print("\n=== Diagn√≥stico inicial de valores categ√≥ricos ===")
for col in colunas_categoricas:
    unique_vals = df_padronizacao[col].unique()
    print(f"\nColuna: {col}")
    print(unique_vals)
    print("Quantidade de valores √∫nicos:", len(unique_vals))

# Verificados valores inconsistentes, temos a necessidade de padroniz√°-los
# Tipo_Eletronico
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].str.strip()  # remove espa√ßos extras
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].replace({'celular': 'Celular'})
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].fillna(df_padronizacao['Tipo_Eletronico'])  # mant√©m NaN

# Origem
df_padronizacao['Origem'] = df_padronizacao['Origem'].str.strip()
df_padronizacao['Origem'] = df_padronizacao['Origem'].replace({'urbano': 'Urbano'})
df_padronizacao['Origem'] = df_padronizacao['Origem'].replace({'Urbana': 'Urbano'})
df_padronizacao['Origem'] = df_padronizacao['Origem'].fillna(df_padronizacao['Origem'])

# Destino_Final
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].str.strip()
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].replace({'reciclagem formal': 'Reciclagem Formal'})
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].fillna(df_padronizacao['Destino_Final'])

# Municipio
df_padronizacao['Municipio'] = df_padronizacao['Municipio'].str.strip()

# Educacao_Ambiental
df_padronizacao['Educacao_Ambiental'] = df_padronizacao['Educacao_Ambiental'].str.strip()

print("\n‚úÖ Valores categ√≥ricos ap√≥s padroniza√ß√£o:")
for col in colunas_categoricas:
    unique_vals = df_padronizacao[col].unique()
    print(f"\nColuna: {col}")
    print(unique_vals)
    print("Quantidade de valores √∫nicos:", len(unique_vals))



print("\n3Ô∏è‚É£ Linhas com valores ausentes\n")

# Seleciona apenas as linhas que possuem pelo menos um valor ausente
linhas_com_na = df_padronizacao[df_padronizacao.isna().any(axis=1)]

# Mostra o total de linhas com valores ausentes
print(f"Total de linhas com pelo menos um valor ausente: {linhas_com_na.shape[0]}\n")

# Exibe as primeiras 10 linhas para confer√™ncia
display(linhas_com_na.head(10))

# Opcional: ver contagem de NaNs por coluna
print("\nüìä Quantidade de valores ausentes por coluna:")
print(df_padronizacao.isna().sum())

print("--- Imputa√ß√£o de valores ausentes ---")

# Colunas num√©ricas
mediana_peso = df_padronizacao['Peso_kg'].median()
mediana_custo = df_padronizacao['Custo_Reciclagem_R$'].median()

# Colunas categ√≥ricas
moda_tipo = df_padronizacao['Tipo_Eletronico'].mode()[0]
moda_origem = df_padronizacao['Origem'].mode()[0]
moda_destino = df_padronizacao['Destino_Final'].mode()[0]

# Aplicando o preenchimento
df_padronizacao['Peso_kg'] = df_padronizacao['Peso_kg'].fillna(mediana_peso)
df_padronizacao['Custo_Reciclagem_R$'] = df_padronizacao['Custo_Reciclagem_R$'].fillna(mediana_custo)
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].fillna(moda_tipo)
df_padronizacao['Origem'] = df_padronizacao['Origem'].fillna(moda_origem)
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].fillna(moda_destino)

# Confer√™ncia final
print("‚úÖ Valores ausentes ap√≥s imputa√ß√£o:")
print(df_padronizacao.isna().sum())

print("\n4Ô∏è‚É£ Verifica√ß√£o de duplicatas ap√≥s padroniza√ß√£o e imputa√ß√£o\n")

# Checamos se existem registros duplicados na base df_padronizacao
duplicatas_encontradas = df_padronizacao.duplicated().sum()

# Informamos o resultado
if duplicatas_encontradas > 0:
    print(f"‚ö†Ô∏è Foram encontradas {duplicatas_encontradas} duplicatas na base de dados.")
    print("As duplicatas podem ser analisadas com df_padronizacao[df_padronizacao.duplicated(keep=False)]")
else:
    print("‚úÖ Nenhuma duplicata encontrada ap√≥s a padroniza√ß√£o e imputa√ß√£o de valores ausentes.")

# Conferir total de registros v√°lidos restantes
print(f"üìä Total de registros v√°lidos: {len(df_padronizacao)}")
```

‚úÖ **Coment√°rios sobre a Atividade 2 ‚Äì Limpeza e Padroniza√ß√£o:**

   - Para facilitar visualiza√ß√£o no padr√£o brasileiro, criamos a coluna **Data_Descarte_BR** no formato `dd/mm/yyyy`.  
   - Isso garante an√°lises temporais consistentes.

2. Padroniza√ß√£o de valores categ√≥ricos:  
   - **Tipo_Eletronico**: registros inconsistentes como `celular` e `Celular` foram unificados em `Celular`.  
   - **Origem**: valores como `Urbana` e `urbano` foram padronizados para `Urbano`.  
   - **Destino_Final**: `reciclagem formal` foi padronizado para `Reciclagem Formal`.  
   - **Municipio** e **Educacao_Ambiental** tiveram apenas limpeza de espa√ßos extras.

3. Valores ausentes:  
   - Foram identificados **64 registros** com pelo menos um valor ausente, afetando principalmente as colunas `Tipo_Eletronico`, `Origem`, `Destino_Final`, `Peso_kg` e `Custo_Reciclagem_R$`.  
   - O preenchimento foi realizado com **mediana** para colunas num√©ricas e **moda** para colunas categ√≥ricas.

4. Duplicatas:  
   - Mesmo ap√≥s a padroniza√ß√£o, n√£o foram identificados registros duplicados.

5. Qualidade geral da base:  
   - Colunas categ√≥ricas consistentes e uniformes.  
   - Colunas num√©ricas sem valores ausentes.

```python
# Usando o DataFrame j√° tratado da Atividade 2

# As importa√ß√µes do seaborn e matplotlib.pyplot foram realizadas na atividade 03, abaixo seguem as adicionais, para essa atividade:
from datetime import datetime
import matplotlib.dates as mdates
import numpy as np


# Verificando novamente a estrutura da coluna Data_Descarte
print("üîç 1. Informa√ß√µes sobre a coluna Data_Descarte:")
print(f"Tipo da coluna: {df_padronizacao['Data_Descarte'].dtype}")
print(f"Primeiros valores:")
display(df_padronizacao[['Data_Descarte', 'Data_Descarte_BR']].head())

# Verificando o per√≠odo abrangido pelos dados
print(f"\nPer√≠odo dos dados:")
print(f"Data mais antiga: {df_padronizacao['Data_Descarte'].min()}")
print(f"Data mais recente: {df_padronizacao['Data_Descarte'].max()}")

# 2. Cria√ß√£o de colunas adicionais para an√°lise temporal
print("\nüìÖ 2. Criando colunas adicionais para an√°lise temporal:")

# Extraindo componentes da data
df_temporal = df_padronizacao.copy()

# Extraindo ano, m√™s e nome do m√™s em portugu√™s
df_temporal['Ano'] = df_temporal['Data_Descarte'].dt.year
df_temporal['Mes'] = df_temporal['Data_Descarte'].dt.month
df_temporal['Nome_Mes'] = df_temporal['Data_Descarte'].dt.strftime('%B').str.capitalize()
df_temporal['Ano_Mes'] = df_temporal['Data_Descarte'].dt.to_period('M')

# Verificando as novas colunas
print("Novas colunas criadas:")
display(df_temporal[['Data_Descarte', 'Ano', 'Mes', 'Nome_Mes', 'Ano_Mes']].head())

print(f"\nResumo temporal:")
print(f"Anos abrangidos: {sorted(df_temporal['Ano'].unique())}")
print(f"Total de registros: {len(df_temporal)}")


# 3. An√°lise de volume de descartes por m√™s/ano
print("\nüìä 3. An√°lise do volume de descartes:")

# Contagem de descartes por per√≠odo (ano-m√™s)
descartes_por_periodo = df_temporal.groupby('Ano_Mes').size().reset_index(name='Quantidade_Descartes')

# Convertendo Ano_Mes para datetime para facilitar a plotagem
descartes_por_periodo['Data_Periodo'] = descartes_por_periodo['Ano_Mes'].dt.to_timestamp()

print("Volume de descartes por per√≠odo:")
display(descartes_por_periodo.head(10))

#Elabora√ß√£o dos gr√°ficos

# Definindo estilo e paleta de cores
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 10})  # Tamanho base da fonte

# LAYOUT OTIMIZADO: Figura maior com espa√ßamentos adequados
fig = plt.figure(figsize=(20, 16))  # Figura ainda maior
gs = fig.add_gridspec(3, 2,
                      height_ratios=[1, 1, 1],
                      width_ratios=[1, 1],
                      hspace=0.35,  # Espa√ßamento vertical maior
                      wspace=0.25,  # Espa√ßamento horizontal adequado
                      left=0.08,    # Margem esquerda
                      right=0.95,   # Margem direita
                      top=0.95,     # Margem superior
                      bottom=0.08)  # Margem inferior

# Criar subplots com layout otimizado
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])

# Gr√°fico 1: Evolu√ß√£o mensal dos descartes

all_periods = pd.period_range(start=df_temporal['Data_Descarte'].min().to_period('M'),
                              end=df_temporal['Data_Descarte'].max().to_period('M'),
                              freq='M')
descartes_por_periodo_all = descartes_por_periodo.set_index('Ano_Mes').reindex(all_periods).fillna(0).reset_index()
descartes_por_periodo_all['Data_Periodo'] = descartes_por_periodo_all['index'].dt.to_timestamp()

ax1.plot(descartes_por_periodo_all['Data_Periodo'],
         descartes_por_periodo_all['Quantidade_Descartes'],
         marker='o', linewidth=2.5, markersize=4, color='salmon', alpha=0.8, markeredgecolor='black')

ax1.set_title('Evolu√ß√£o Mensal do Volume de Descartes', fontsize=14, fontweight='bold', pad=14)
ax1.set_xlabel('Per√≠odo', fontsize=10)
ax1.set_ylabel('Quantidade de Descartes', fontsize=10)
ax1.grid(True, alpha=0.3, linestyle='--')
ax1.tick_params(axis='x', rotation=45, labelsize=9)
ax1.tick_params(axis='y', labelsize=9)
ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))


# Destacando pico m√°ximo com posicionamento inteligente
max_descarte = descartes_por_periodo_all.loc[descartes_por_periodo_all['Quantidade_Descartes'].idxmax()]
ax1.annotate(f'Pico: {int(max_descarte["Quantidade_Descartes"])}',
             xy=(max_descarte['Data_Periodo'], max_descarte['Quantidade_Descartes']),
             xytext=(15, 15), textcoords='offset points',
             bbox=dict(boxstyle='round,pad=0.5', facecolor='red', alpha=0.8, edgecolor='black', linewidth=1),
             arrowprops=dict(arrowstyle='->', color='red', lw=1.0),
             fontsize=7, fontweight='bold')

# Gr√°fico 2: Volume por ano
descartes_por_ano = df_temporal.groupby('Ano').size().reset_index(name='Quantidade_Descartes')
bars = ax2.bar(descartes_por_ano['Ano'], descartes_por_ano['Quantidade_Descartes'],
               color=sns.color_palette("Blues_d", 1)[0], alpha=0.8, edgecolor='black', linewidth=1)

ax2.set_title('Volume Total de Descartes por Ano', fontsize=14, fontweight='bold', pad=15)
ax2.set_xlabel('Ano', fontsize=11)
ax2.set_ylabel('Quantidade de Descartes', fontsize=11)
ax2.grid(True, alpha=0.3, axis='y', linestyle='--')
ax2.set_xticks(range(2015, 2026))


# Valores nas barras com posicionamento otimizado
for bar, value in zip(bars, descartes_por_ano['Quantidade_Descartes']):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{value}', ha='center', va='bottom', fontweight='bold', fontsize=10)

# Gr√°fico 3: Distribui√ß√£o mensal (todos os anos)
descartes_por_mes = df_temporal.groupby('Mes').size().reset_index(name='Quantidade')

# Nomes dos meses em portugu√™s
meses_nomes_pt = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
                  'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

# Plotar como gr√°fico de linhas
ax3.plot(descartes_por_mes['Mes'], descartes_por_mes['Quantidade'], marker='o', linewidth=2.5, markersize=5, color='forestgreen')

ax3.set_title('Distribui√ß√£o de Descartes por M√™s (Todos os Anos)', fontsize=14, fontweight='bold', pad=15)
ax3.set_xlabel('M√™s', fontsize=11)
ax3.set_ylabel('Total de Descartes', fontsize=11)
ax3.set_xticks(range(1, 13))
ax3.set_xticklabels(meses_nomes_pt, fontsize=9)
ax3.tick_params(axis='y', labelsize=9)
ax3.grid(True, alpha=0.3, axis='y', linestyle='--')

# 4. Estat√≠sticas descritivas da an√°lise temporal
print("\nüìã 4. Estat√≠sticas temporais:")

# Estat√≠sticas b√°sicas
estatisticas_mensais = descartes_por_periodo_all['Quantidade_Descartes'].describe().to_frame('Estat√≠sticas do Volume Mensal')

print("Estat√≠sticas do volume mensal de descartes:")
display(estatisticas_mensais)

# Per√≠odo com maior/menor atividade
periodo_max = descartes_por_periodo_all.loc[descartes_por_periodo_all['Quantidade_Descartes'].idxmax()]
periodo_min = descartes_por_periodo_all.loc[descartes_por_periodo_all['Quantidade_Descartes'].idxmin()]

print(f"\nPer√≠odo com MAIOR volume de descartes:")
print(f"  {periodo_max['index']}: {int(periodo_max['Quantidade_Descartes'])} descartes")

print(f"\nPer√≠odo com MENOR volume de descartes:")
print(f"  {periodo_min['index']}: {int(periodo_min['Quantidade_Descartes'])} descartes")

# An√°lise de tend√™ncia simples
print(f"\nTend√™ncia geral:")
media_primeira_metade = descartes_por_periodo_all.head(len(descartes_por_periodo_all)//2)['Quantidade_Descartes'].mean()
media_segunda_metade = descartes_por_periodo_all.tail(len(descartes_por_periodo_all)//2)['Quantidade_Descartes'].mean()

if media_segunda_metade > media_primeira_metade:
    tendencia = "CRESCENTE"
else:
    tendencia = "DECRESCENTE"

print(f"  Primeira metade do per√≠odo: {media_primeira_metade:.1f} descartes/m√™s em m√©dia")
print(f"  Segunda metade do per√≠odo: {media_segunda_metade:.1f} descartes/m√™s em m√©dia")
print(f"  Tend√™ncia: {tendencia}")


print("‚úÖ An√°lise temporal conclu√≠da!")

# 5. Gr√°ficos de evolu√ß√£o temporal
print("\nüìà 5. Gr√°ficos de evolu√ß√£o temporal:")
```

# **:: Atividade 5 ‚Äì Agrega√ß√£o por Munic√≠pio**

<h4 align="justify">A gest√£o de res√≠duos eletr√¥nicos exige que munic√≠pios compreendam seu papel na gera√ß√£o e no destino do lixo tecnol√≥gico.</h4>
<h4 align="justify">Utilize groupby para agregar os dados por munic√≠pio, calculando:</h4>
<h4 align="justify">&bullet; o total de res√≠duos descartados;</h4>
<h4 align="justify">&bullet; a m√©dia de peso dos eletr√¥nicos; e</h4>
<h4 align="justify">&bullet; e a propor√ß√£o de destinos formais (reciclagem, doa√ß√£o) em rela√ß√£o aos informais (lix√£o, descarte irregular).</h4>
<h4 align="justify">Apresente os resultados em uma tabela ordenada e em gr√°fico de barras horizontais.</h4>

```python
# 1. Criando c√≥pia do dataset tratado para manipula√ß√£o nesta atividade
df_atv5 = df_padronizacao.copy()

# 2. Agregar os dados por 'Municipio'
total_residuos = df_atv5.groupby('Municipio').size().rename('Total_Residuos')
media_peso = df_atv5.groupby('Municipio')['Peso_kg'].mean().rename('Media_Peso_kg')

# Todos os destinos finais: ['Aterro Controlado' 'Cooperativa' 'Reciclagem Formal' 'Lix√£o' 'Rio']
destinos_formais = ['Cooperativa', 'Reciclagem Formal']
df_atv5['Tipo_Destino'] = df_atv5['Destino_Final'].apply(lambda x: 'Formal' if x in destinos_formais else 'Informal')
contagem_destinos = df_atv5.groupby(['Municipio', 'Tipo_Destino']).size().unstack(fill_value=0)
proporcao_formal = (contagem_destinos.get('Formal', 0) / (contagem_destinos.get('Formal', 0) + contagem_destinos.get('Informal', 0))).rename('Proporcao_Formal')

# 3. Combinar e ordenar a tabela
tabela_agregada = pd.concat([total_residuos, media_peso, proporcao_formal], axis=1).fillna(0)
tabela_agregada = tabela_agregada.sort_values(by='Total_Residuos', ascending=False)

# CORRE√á√ÉO DA TABELA: Formatar e imprimir
print("--- Tabela Agregada por Munic√≠pio ---")
tabela_formatada = tabela_agregada.copy()
tabela_formatada['Media_Peso_kg'] = tabela_formatada['Media_Peso_kg'].map('{:.2f}'.format)
tabela_formatada['Proporcao_Formal'] = tabela_formatada['Proporcao_Formal'].map('{:.2%}'.format)
print(tabela_formatada)

# 4. Apresenta√ß√£o em Gr√°ficos de Barras Horizontais
fig, axes = plt.subplots(3, 1, figsize=(10, 15))
fig.suptitle('An√°lise Agregada por Munic√≠pio', fontsize=16)

# Gr√°fico 1: Total de Res√≠duos
sns.barplot(x=tabela_agregada['Total_Residuos'], y=tabela_agregada.index, hue=tabela_agregada.index, ax=axes[0], palette='viridis', legend=False)
axes[0].set_title('Total de Res√≠duos Descartados por Munic√≠pio')
axes[0].set_xlabel('Quantidade de Itens')
axes[0].set_ylabel('Munic√≠pio')

# Gr√°fico 2: M√©dia de Peso
sns.barplot(x=tabela_agregada['Media_Peso_kg'], y=tabela_agregada.index, hue=tabela_agregada.index, ax=axes[1], palette='plasma', legend=False)
axes[1].set_title('M√©dia de Peso (kg) dos Eletr√¥nicos por Munic√≠pio')
axes[1].set_xlabel('Peso M√©dio (kg)')
axes[1].set_ylabel('Munic√≠pio')

# Gr√°fico 3: Propor√ß√£o de Destinos Formais
sns.barplot(x=tabela_agregada['Proporcao_Formal'], y=tabela_agregada.index, hue=tabela_agregada.index, ax=axes[2], palette='magma', legend=False)
axes[2].set_title('Propor√ß√£o de Descarte em Destinos Formais por Munic√≠pio')
axes[2].set_xlabel('Propor√ß√£o (%)')
axes[2].set_ylabel('Munic√≠pio')
axes[2].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0%}'.format(x)))

plt.tight_layout()
plt.show()
```

3. **Munic√≠pio e Propor√ß√£o de Descartes em Destinos Formais**
    
    Destinos formais: Cooperativa e Reciclagem Formal.

# **:: Atividade 6 - Cruzamento entre Tipo de Eletr√¥nico e Destino**

<h4 align="justify">Nem todos os eletr√¥nicos t√™m o mesmo destino: alguns aparelhos t√™m maior chance de serem reciclados, enquanto outros acabam em descartes irregulares.</h4>
<h4 align="justify">Crie uma tabela cruzada (crosstab ou pivot_table) relacionando o tipo de eletr√¥nico com seu destino. Visualize os resultados em um heatmap (Seaborn), destacando quais combina√ß√µes s√£o mais cr√≠ticas.</h4>

```python
df_padronizacao_copia = df_padronizacao.copy() # para manuten√ß√£o da integridade do que j√° foi tratado, vamos utilizar uma c√≥pia

# 1. Criar a Tabela Cruzada (Crosstab)

tabela_cruzada = pd.crosstab(df_padronizacao_copia['Tipo_Eletronico'], df_padronizacao_copia['Destino_Final']) # quantidade em inteiro
tabela_cruzada.columns.name = None # Remover o nome da coluna Destino_Final

tabela_percentual = pd.crosstab(df_padronizacao_copia['Tipo_Eletronico'], df_padronizacao_copia['Destino_Final'], normalize='index') * 100 #traz a distribui√ß√£o em percentual
tabela_percentual_formatada = tabela_percentual.style.format('{:.2f}%') # formata√ß√£o para aparecer na tabela
tabela_percentual_formatada.columns.name = None # Remover o nome da coluna Destino_Final


print("--- Tabela Cruzada: Tipo de Eletr√¥nico vs. Destino Final ---")
display(tabela_cruzada)

print("\n\n--- Tabela Cruzada: Tipo de Eletr√¥nico vs. Destino Final (%) ---")
display(tabela_percentual_formatada)


# 2. Visualizar os resultados em um Heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(
    tabela_cruzada,
    annot=True,       # mostra os valores nas c√©lulas
    cmap="YlOrRd",    # paleta de cores amarelo, laranja e vermelho, para destacarem pontos cr√≠ticos
    fmt='g'           # formato dos n√∫meros (g = inteiro)
)

print("\n--- Heatmap: Tipo de Eletr√¥nico vs. Destino Final ---\n")
plt.title('Rela√ß√£o entre Tipo de Eletr√¥nico e seu Destino Final', fontsize=24)
plt.xlabel('Destino Final', fontsize=12)
plt.ylabel('Tipo de Eletr√¥nico', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

#tamb√©m em percentual de eletr√¥nico por destino final

plt.figure(figsize=(14, 8))
sns.heatmap(
    tabela_percentual,
    annot=True,       # mostra os valores nas c√©lulas
    cmap="Oranges",    # paleta de tons alaranjados, para destacarem pontos cr√≠ticos
    fmt='.2f'           # formato dos n√∫meros (f = float)
)

print("\n--- Heatmap: Tipo de Eletr√¥nico vs. Destino Final (%) ---\n")
plt.title('Rela√ß√£o entre Tipo de Eletr√¥nico e seu Destino Final', fontsize=24)
plt.xlabel('Destino Final', fontsize=12)
plt.ylabel('Tipo de Eletr√¥nico', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

# **:: Atividade 7 ‚Äì Impacto do Peso dos Res√≠duos no Meio Ambiente**

<h4 align="justify">O peso dos res√≠duos eletr√¥nicos √© um fator cr√≠tico para a log√≠stica de coleta e o impacto ambiental. Equipamentos mais pesados, como geladeiras e televisores, podem sobrecarregar aterros irregulares e dificultar o transporte para centros de reciclagem. Na Amaz√¥nia, onde a infraestrutura de estradas √© limitada, esse aspecto √© ainda mais sens√≠vel.</h4>
<h4 align="justify">Al√©m disso, a presen√ßa de metais pesados em equipamentos volumosos pode gerar contamina√ß√µes irrevers√≠veis em rios e no solo. Analisar a distribui√ß√£o do peso √© fundamental para indicar se h√° tend√™ncia de ac√∫mulo de grandes aparelhos ou de pequenos dispositivos que passam despercebidos, mas se espalham de forma descontrolada.</h4>
<h4 align="justify">Utilize describe() e m√©todos de visualiza√ß√£o (boxplot e histplot) para analisar a distribui√ß√£o do peso dos res√≠duos eletr√¥nicos. Identifique outliers e discuta poss√≠veis explica√ß√µes para sua ocorr√™ncia.</h4>
<h4 align="justify">Em seguida, fa√ßa uma compara√ß√£o da m√©dia e do desvio padr√£o de peso entre descartes destinados formalmente e informalmente, apresentando os resultados em gr√°fico de violino.</h4>

```python
# Para responder essa quest√£o, √© necess√°rio o importe do pandas, seaborn e matplotlib.pyplot, que j√° foram realizados anteriormente

df_padronizacao_copia = df_padronizacao.copy() # para manuten√ß√£o da integridade do que j√° foi tratado, vamos utilizar uma c√≥pia

# 1. An√°lise da Distribui√ß√£o do Peso (Peso_kg)
print("--- 1. An√°lise Descritiva da Coluna 'Peso_kg' ---")
print(df_padronizacao_copia['Peso_kg'].describe())

# Gr√°ficos para visualizar a distribui√ß√£o
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Distribui√ß√£o do Peso (kg) dos Res√≠duos Eletr√¥nicos', fontsize=16)

# Boxplot para identificar outliers
sns.boxplot(y=df_padronizacao_copia['Peso_kg'], ax=axes[0])
axes[0].set_title('Boxplot do Peso')
axes[0].set_ylabel('Peso (kg)')


# Histograma para ver a frequ√™ncia
sns.histplot(df_padronizacao_copia['Peso_kg'],kde=True, ax=axes[1])
plt.title('Histograma do Peso')
plt.xlabel('Peso (kg)')
plt.ylabel('Frequ√™ncia')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# 2. Identifica√ß√£o Autom√°tica de Outliers (IQR)
Q1 = df_padronizacao_copia['Peso_kg'].quantile(0.25)
Q3 = df_padronizacao_copia['Peso_kg'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df_padronizacao_copia[(df_padronizacao_copia['Peso_kg'] < limite_inferior) | (df_padronizacao_copia['Peso_kg'] > limite_superior)]

print("\n--- 2. Identifica√ß√£o de Outliers ---")
if outliers.empty:
    print(""" Os gr√°ficos mostram que a maior parte dos res√≠duos eletr√¥nicos pesa entre 20 e 30 kg, com uma m√©dia pr√≥xima de 25 kg.
    N√£o foram encontrados outliers, o que indica que os dados est√£o consistentes e sem valores muito fora do padr√£o.
    Ainda assim, aparecem alguns itens mais pesados, chegando perto de 50 kg, provavelmente geladeiras ou televisores,
    o que √© esperado considerando o tipo de material descartado.""")


# 3. Compara√ß√£o entre Destinos Formais e Informais
destinos_formais = ['Reciclagem Formal', 'Cooperativa']
df_padronizacao_copia['Tipo_Destino'] = df_padronizacao_copia['Destino_Final'].apply(
    lambda x: 'Formal' if x in destinos_formais else 'Informal'
)
estatisticas_peso_destino = df_padronizacao_copia.groupby('Tipo_Destino')['Peso_kg'].agg(['mean', 'std'])

print("\n--- 3. M√©dia e Desvio Padr√£o do Peso por Tipo de Destino ---")
print(estatisticas_peso_destino)

# Gr√°fico de Violino (corrigido, sem duplica√ß√£o de categorias)
plt.figure(figsize=(10, 7))
sns.violinplot(x='Tipo_Destino', y='Peso_kg', data=df_padronizacao_copia, palette='muted')

plt.title('Distribui√ß√£o do Peso por Tipo de Destino (Formal vs. Informal)', fontsize=16)
plt.xlabel('Tipo de Destino', fontsize=12)
plt.ylabel('Peso (kg)', fontsize=12)
plt.show()
```

# **:: Atividade 8 ‚Äì Rela√ß√£o entre Educa√ß√£o Ambiental e Destino do Res√≠duo**

<h4 align="justify">Um dos fatores determinantes para o destino correto dos res√≠duos eletr√¥nicos √© o n√≠vel de conscientiza√ß√£o ambiental da popula√ß√£o. Munic√≠pios que aplicam programas de educa√ß√£o ambiental tendem a apresentar maior propor√ß√£o de descarte formal, reduzindo a polui√ß√£o gerada por lix√µes clandestinos.</h4>
<h4 align="justify">No entanto, essa rela√ß√£o nem sempre √© linear. √Äs vezes, mesmo com iniciativas locais, a aus√™ncia de infraestrutura de coleta inviabiliza o descarte correto. A an√°lise comparativa pode revelar se a educa√ß√£o ambiental tem impacto mensur√°vel no destino dos res√≠duos registrados na base.</h4>
<h4 align="justify">Agrupe os dados por munic√≠pio, comparando o percentual de descarte formal em locais com e sem registro de programas de educa√ß√£o ambiental.</h4>
<h4 align="justify">Crie gr√°ficos de barras empilhadas mostrando a propor√ß√£o de destinos em cada categoria, e discuta os poss√≠veis vieses do dataset (ex.: munic√≠pios com dados incompletos).</h4>

```python
import matplotlib.ticker as mtick # Para formatar eixos em porcentagem

# Assumindo que 'df_padronizacao' √© o DataFrame limpo da atividade anterior
# Criamos uma c√≥pia para esta atividade para garantir a integridade dos dados
df_ed_amb = df_padronizacao.copy(deep=True)


# ==============================================================================
# 1. AN√ÅLISE PRIM√ÅRIA (Resposta Direta √† Tarefa)
# ==============================================================================
print("\n" + "="*80)
print("--- 1. AN√ÅLISE PRIM√ÅRIA: Propor√ß√£o de Destinos por Programa de Educa√ß√£o ---")
print("="*80)

# Agrupar por 'Educacao_Ambiental' e 'Destino_Final' e calcular as propor√ß√µes
proporcoes_destinos = df_ed_amb.groupby('Educacao_Ambiental')['Destino_Final'] \
                        .value_counts(normalize=True) \
                        .mul(100) \
                        .rename('Percentual') \
                        .unstack(fill_value=0)

print("\nTabela de propor√ß√µes de cada destino (%) por categoria de programa:")
print(proporcoes_destinos.to_string(float_format="%.2f%%"))

# Gerar o gr√°fico de barras empilhadas principal
ax = proporcoes_destinos.plot(
    kind='barh',
    stacked=True,
    figsize=(12, 6),
    colormap='viridis'
)
plt.title('Propor√ß√£o de Destinos Finais por Programa de Educa√ß√£o Ambiental', fontsize=16)
plt.xlabel('Percentual (%)', fontsize=12)
plt.ylabel('Exist√™ncia de Programa de Educa√ß√£o', fontsize=12)
plt.legend(title='Destino Final', bbox_to_anchor=(1.02, 1), loc='upper left')
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
for c in ax.containers:
    ax.bar_label(c, label_type='center', fmt='%.1f%%', color='white', weight='bold')
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()


# ==============================================================================
# 2. AN√ÅLISE SECUND√ÅRIA (Investiga√ß√£o de Vieses e Fatores Contextuais)
# ==============================================================================
print("\n" + "="*80)
print("--- 2. AN√ÅLISE SECUND√ÅRIA: INVESTIGA√á√ÉO DE VIESES E FATORES CONTEXTUAIS ---")
print("="*80)

# --- ETAPA A: PREPARA√á√ÉO DOS DADOS E IMPRESS√ÉO DAS AN√ÅLISES TEXTUAIS ---
destinos_formais = ['Aterro controlado', 'Cooperativa', 'Reciclagem formal']
df_ed_amb['Tipo_Destino'] = df_ed_amb['Destino_Final'].apply(lambda x: 'Formal' if x in destinos_formais else 'Informal')

# 2.1 An√°lise de Vi√©s de Sele√ß√£o
print("\n--- 2.1. Investiga√ß√£o de Vi√©s de Sele√ß√£o ---\n")
contagem_por_municipio = df_ed_amb['Municipio'].value_counts()
print("Resumo: A contagem de registros por munic√≠pio √© altamente desigual.")
print("Poucos munic√≠pios (ex: Bel√©m, Ananindeua) dominam o dataset, o que pode distorcer a m√©dia geral.")
print("\nTabela: Contagem de Registros por Munic√≠pio")
print(contagem_por_municipio.to_frame(name='Total_Registros').head())

# 2.2 An√°lise de Vi√©s de Defini√ß√£o
print("\n" + "-"*80)
print("\n--- 2.2. Investiga√ß√£o de Vi√©s de Defini√ß√£o ---\n")
df_belem_sim = df_ed_amb[(df_ed_amb['Municipio'] == 'Bel√©m') & (df_ed_amb['Educacao_Ambiental'] == 'Sim')]
desempenho_belem = df_belem_sim['Destino_Final'].value_counts(normalize=True).mul(100)
print("Resumo: A etiqueta 'Educa√ß√£o Sim' √© question√°vel. Bel√©m, um munic√≠pio com programa,")
print(f"apresenta uma taxa de descarte informal (Lix√£o + Rio) de {desempenho_belem.get('Lix√£o', 0) + desempenho_belem.get('Rio', 0):.1f}%.")
print("\nTabela: Propor√ß√£o de Destinos em Bel√©m (Com Programa de Educa√ß√£o)")
print(desempenho_belem.to_frame(name='Percentual (%)').to_string(float_format="%.2f%%"))

# 2.3 An√°lise da Rela√ß√£o entre Educa√ß√£o e Origem
print("\n" + "-"*80)
print("\n--- 2.3. An√°lise da Rela√ß√£o entre Educa√ß√£o e Origem dos Res√≠duos ---\n")
df_origem_proporcoes = df_ed_amb.groupby(['Educacao_Ambiental', 'Origem'])['Tipo_Destino'].value_counts(normalize=True).mul(100).rename('Percentual').reset_index()
df_formal_por_origem = df_origem_proporcoes[df_origem_proporcoes['Tipo_Destino'] == 'Formal']
# <-- ALTERA√á√ÉO AQUI: Deixando a defini√ß√£o expl√≠cita no resumo textual -->
print("Resumo: O impacto da educa√ß√£o ambiental varia por setor. Para esta an√°lise, o Descarte Formal")
print("considera as categorias: 'Aterro controlado', 'Cooperativa' e 'Reciclagem formal'.")
print("O √∫nico aumento significativo ocorre no setor Industrial (de 9.1% para 17.6%).")
print("\nTabela: Percentual de Descarte Formal por Origem e Programa de Educa√ß√£o")
print(df_formal_por_origem[['Educacao_Ambiental', 'Origem', 'Percentual']].to_string(index=False, float_format="%.2f%%"))

# 2.4 An√°lise de Outliers ("Her√≥is" e "Vil√µes")
print("\n" + "-"*80)
print("\n--- 2.4. An√°lise de Outliers: 'Her√≥is' e 'Vil√µes' ---\n")
# <-- ALTERA√á√ÉO AQUI: Deixando a defini√ß√£o expl√≠cita no resumo textual -->
print("Resumo: Munic√≠pios 'Her√≥is' (sem programa) com 100% de descarte formal sugerem que a")
print("infraestrutura pode ser mais crucial. 'Vil√µes' (com programa) com baixo desempenho confirmam")
print("que a educa√ß√£o, sozinha, n√£o √© garantia de sucesso. (Descarte Formal = Aterro, Cooperativa, Reciclagem).")
MINIMO_REGISTROS = 3
taxas_formais = df_ed_amb[df_ed_amb['Tipo_Destino'] == 'Formal'].groupby('Municipio').size()
total_registros = df_ed_amb.groupby('Municipio').size()
proporcao_formal_municipio = (taxas_formais / total_registros * 100).fillna(0)
info_municipios = df_ed_amb.groupby('Municipio').agg(Educacao_Ambiental=('Educacao_Ambiental', 'first'), Total_Registros=('ID_Registro', 'count')).join(proporcao_formal_municipio.rename('Percentual_Formal')).reset_index()
analise_confiavel = info_municipios[info_municipios['Total_Registros'] >= MINIMO_REGISTROS]
herois = analise_confiavel[analise_confiavel['Educacao_Ambiental'] == 'N√£o'].sort_values(by='Percentual_Formal', ascending=False)
viloes = analise_confiavel[analise_confiavel['Educacao_Ambiental'] == 'Sim'].sort_values(by='Percentual_Formal', ascending=True)
df_outliers_plot = pd.concat([herois.head(3), viloes.head(3)])
df_outliers_plot['Tipo'] = ['Her√≥i'] * 3 + ['Vil√£o'] * 3
print("\n'Her√≥is' (Top 3):")
print(herois[['Municipio', 'Percentual_Formal', 'Total_Registros']].head(3).to_string(index=False, float_format="%.2f%%"))
print("\n'Vil√µes' (Top 3):")
print(viloes[['Municipio', 'Percentual_Formal', 'Total_Registros']].head(3).to_string(index=False, float_format="%.2f%%"))

# --- ETAPA B: GERA√á√ÉO DO PAINEL VISUAL CONSOLIDADO ---
print("\n\n" + "="*80)
print("--- PAINEL VISUAL CONSOLIDADO DAS AN√ÅLISES SECUND√ÅRIAS ---")
print("="*80)

fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Painel de An√°lise de Vieses e Fatores Contextuais', fontsize=20, weight='bold')
ax1, ax2, ax3, ax4 = axes.flatten()

# Gr√°fico 2.1
sns.barplot(x=contagem_por_municipio.index, y=contagem_por_municipio.values, hue=contagem_por_municipio.index, palette='crest', ax=ax1, legend=False)
ax1.set_title('2.1. Vi√©s de Sele√ß√£o: Registros por Munic√≠pio', fontsize=14)
ax1.set_xlabel(''), ax1.set_ylabel('N√∫mero de Registros')
ax1.tick_params(axis='x', rotation=45, labelsize=9)

# Gr√°fico 2.2
sns.barplot(x=desempenho_belem.index, y=desempenho_belem.values, hue=desempenho_belem.index, palette='magma', ax=ax2, legend=False)
ax2.set_title("2.2. Vi√©s de Defini√ß√£o: Desempenho de Bel√©m", fontsize=14)
ax2.set_xlabel('Destino Final'), ax2.set_ylabel('Percentual (%)')
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())

# Gr√°fico 2.3
sns.barplot(data=df_formal_por_origem, x='Origem', y='Percentual', hue='Educacao_Ambiental', palette='rocket', ax=ax3)
ax3.set_title('2.3. Impacto da Educa√ß√£o por Origem do Res√≠duo', fontsize=14)
# <-- ALTERA√á√ÉO AQUI: Deixando a defini√ß√£o expl√≠cita no eixo y do gr√°fico -->
ax3.set_xlabel('Origem do Res√≠duo'), ax3.set_ylabel('Descarte Formal (%)\n(Aterro, Cooperativa, Reciclagem)', fontsize=12)
ax3.set_ylim(0, 100), ax3.yaxis.set_major_formatter(mtick.PercentFormatter())

# Gr√°fico 2.4
sns.barplot(data=df_outliers_plot, x='Municipio', y='Percentual_Formal', hue='Tipo', palette={'Her√≥i': 'green', 'Vil√£o': 'red'}, ax=ax4)
ax4.set_title('2.4. Outliers de Desempenho ("Her√≥is" vs. "Vil√µes")', fontsize=14)
# <-- ALTERA√á√ÉO AQUI: Deixando a defini√ß√£o expl√≠cita no eixo y do gr√°fico -->
ax4.set_xlabel('Munic√≠pio'), ax4.set_ylabel('Descarte Formal (%)\n(Aterro, Cooperativa, Reciclagem)', fontsize=12)
ax4.set_ylim(0, 100), ax4.yaxis.set_major_formatter(mtick.PercentFormatter())
ax4.tick_params(axis='x', rotation=45, labelsize=9)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# **:: Atividade 9 ‚Äì Correla√ß√£o entre Origem e Tipo de Eletr√¥nico**

<h4 align="justify">A origem do res√≠duo eletr√¥nico pode estar associada ao tipo de dispositivo descartado. Grandes centros urbanos tendem a gerar computadores, smartphones e notebooks, enquanto √°reas rurais podem descartar r√°dios, televisores antigos e baterias automotivas. Essa correla√ß√£o ajuda a entender os padr√µes de consumo e de descarte.</h4>
<h4 align="justify">Compreender tais rela√ß√µes √© vital para desenhar pol√≠ticas regionais: por exemplo, pontos de coleta de baterias em cidades pequenas ou centrais de reciclagem de computadores em polos universit√°rios. Identificar as vari√°veis que caminham juntas pode revelar gargalos e oportunidades estrat√©gicas.</h4>
<h4 align="justify">Calcule a correla√ß√£o entre origem (urbana/rural/industrial) e tipo de eletr√¥nico usando an√°lise de conting√™ncia (qui-quadrado).</h4>
<h4 align="justify">Calcule a correla√ß√£o entre origem (urbana/rural/industrial) e tipo de eletr√¥nico usando an√°lise de conting√™ncia (qui-quadrado).</h4>

```python
from scipy.stats import chi2_contingency

df_padronizacao_copia = df_padronizacao.copy() # para manuten√ß√£o da integridade do que j√° foi tratado, vamos utilizar uma c√≥pia

# 1. Criar a Tabela de Conting√™ncia (Crosstab)
tabela_contingencia = pd.crosstab(df_padronizacao_copia['Origem'], df_padronizacao_copia['Tipo_Eletronico'])

print("--- Tabela de Conting√™ncia: Origem vs. Tipo de Eletr√≥nico ---")
print(tabela_contingencia)


# 2. Realizar o Teste Qui-Quadrado
chi2, p_value, dof, expected = chi2_contingency(tabela_contingencia)

print("\n--- Resultados do Teste Qui-Quadrado ---")
print(f"Estat√≠stica Qui-Quadrado (œá¬≤): {chi2:.4f}")
print(f"Valor-p (p-value): {p_value:.4f}")
print(f"Graus de Liberdade (dof): {dof}")


# 3. Interpreta√ß√£o do Resultado
print("\n--- Interpreta√ß√£o ---")
alpha = 0.05  # N√≠vel de signific√¢ncia comum

if p_value < alpha:
    print(f"Como o valor-p ({p_value:.4f}) √© menor que {alpha}, rejeitamos a hip√≥tese nula.")
    print("Conclus√£o: Existe uma associa√ß√£o estatisticamente significativa entre a Origem do res√≠duo e o Tipo de Eletr√≥nico.")
else:
    print(f"Como o valor-p ({p_value:.4f}) n√£o √© menor que {alpha}, n√£o podemos rejeitar a hip√≥tese nula.")
    print("Conclus√£o: N√£o h√° evid√™ncias de uma associa√ß√£o estatisticamente significativa entre a Origem do res√≠duo e o Tipo de Eletr√≥nico.")
# 1. Heatmap da tabela de conting√™ncia
plt.figure(figsize=(10,6))
sns.heatmap(tabela_contingencia, annot=True, cmap="Blues", fmt="d")
plt.title("Heatmap - Origem vs. Tipo de Eletr√¥nico")
plt.ylabel("Origem")
plt.xlabel("Tipo de Eletr√¥nico")
plt.show()

# 2. Gr√°fico de barras agrupadas
tabela_contingencia.plot(kind="bar", figsize=(10,6))
plt.title("Distribui√ß√£o dos Tipos de Eletr√¥nicos por Origem")
plt.ylabel("Frequ√™ncia")
plt.xlabel("Origem")
plt.xticks(rotation=0)
plt.legend(title="Tipo de Eletr√¥nico")
plt.show()
```

*   N√£o existe evid√™ncia forte de que, por exemplo, "o meio rural descarta mais tablets que o urbano" ou "o governo descarta mais geladeiras que as ind√∫strias".
