# **:: Atividade 2 – Limpeza e Padronização**

<h4 align="justify">A diversidade de origens e destinos pode gerar confusões quando informações semelhantes aparecem de forma inconsistente, prejudicando relatórios ambientais e logísticos.</h4>
<h4 align="justify">Identifique e trate valores inconsistentes (ex.: "reciclagem formal" e "Reciclagem Formal"), corrija erros de digitação e padronize todas as colunas categóricas. Em seguida, elimine duplicatas mantendo apenas registros válidos.</h4>

```python
# Instala a biblioteca unicode no Colab
!pip install unidecode

# OBS: Pandas já foi importado na Atividade 1
from unidecode import unidecode
import re

# Fazemos uma cópia da DB para fazer a padronização
df_padronizacao = df.copy()

print("🔄 Limpeza e Padronização\n")

print("1️⃣ Conversão da coluna Data_Descarte para datetime")
# Transformamos de object para datetime e depois mantemos apenas a parte da data em si
# Tratamos possíveis formatos inconsistentes
df_padronizacao["Data_Descarte"] = pd.to_datetime(df_padronizacao["Data_Descarte"], errors="coerce")

# Conferir os primeiros valores para validar a conversão
print("✅ Primeiras linhas da coluna Data_Descarte após conversão:")
display(df_padronizacao[["Data_Descarte"]].head(10))

# Criar coluna formatada para visualização em formato brasileiro
df_padronizacao['Data_Descarte_BR'] = df_padronizacao['Data_Descarte'].dt.strftime('%d/%m/%Y')

# Conferir primeiras linhas
display(df_padronizacao[['Data_Descarte', 'Data_Descarte_BR']].head(10))


print("\n2️⃣ Diagnóstico inicial dos valores categóricos")
colunas_categoricas = ["Tipo_Eletronico", "Origem", "Destino_Final", "Municipio", "Educacao_Ambiental"]

print("\n=== Diagnóstico inicial de valores categóricos ===")
for col in colunas_categoricas:
    unique_vals = df_padronizacao[col].unique()
    print(f"\nColuna: {col}")
    print(unique_vals)
    print("Quantidade de valores únicos:", len(unique_vals))

# Verificados valores inconsistentes, temos a necessidade de padronizá-los
# Tipo_Eletronico
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].str.strip()  # remove espaços extras
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].replace({'celular': 'Celular'})
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].fillna(df_padronizacao['Tipo_Eletronico'])  # mantém NaN

# Origem
df_padronizacao['Origem'] = df_padronizacao['Origem'].str.strip()
df_padronizacao['Origem'] = df_padronizacao['Origem'].replace({'urbano': 'Urbano'})
df_padronizacao['Origem'] = df_padronizacao['Origem'].replace({'Urbana': 'Urbano'})
df_padronizacao['Origem'] = df_padronizacao['Origem'].fillna(df_padronizacao['Origem'])

# Destino_Final
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].str.strip()
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].replace({'reciclagem formal': 'Reciclagem Formal'})
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].fillna(df_padronizacao['Destino_Final'])

# Municipio
df_padronizacao['Municipio'] = df_padronizacao['Municipio'].str.strip()

# Educacao_Ambiental
df_padronizacao['Educacao_Ambiental'] = df_padronizacao['Educacao_Ambiental'].str.strip()

print("\n✅ Valores categóricos após padronização:")
for col in colunas_categoricas:
    unique_vals = df_padronizacao[col].unique()
    print(f"\nColuna: {col}")
    print(unique_vals)
    print("Quantidade de valores únicos:", len(unique_vals))



print("\n3️⃣ Linhas com valores ausentes\n")

# Seleciona apenas as linhas que possuem pelo menos um valor ausente
linhas_com_na = df_padronizacao[df_padronizacao.isna().any(axis=1)]

# Mostra o total de linhas com valores ausentes
print(f"Total de linhas com pelo menos um valor ausente: {linhas_com_na.shape[0]}\n")

# Exibe as primeiras 10 linhas para conferência
display(linhas_com_na.head(10))

# Opcional: ver contagem de NaNs por coluna
print("\n📊 Quantidade de valores ausentes por coluna:")
print(df_padronizacao.isna().sum())

print("--- Imputação de valores ausentes ---")

# Colunas numéricas
mediana_peso = df_padronizacao['Peso_kg'].median()
mediana_custo = df_padronizacao['Custo_Reciclagem_R$'].median()

# Colunas categóricas
moda_tipo = df_padronizacao['Tipo_Eletronico'].mode()[0]
moda_origem = df_padronizacao['Origem'].mode()[0]
moda_destino = df_padronizacao['Destino_Final'].mode()[0]

# Aplicando o preenchimento
df_padronizacao['Peso_kg'] = df_padronizacao['Peso_kg'].fillna(mediana_peso)
df_padronizacao['Custo_Reciclagem_R$'] = df_padronizacao['Custo_Reciclagem_R$'].fillna(mediana_custo)
df_padronizacao['Tipo_Eletronico'] = df_padronizacao['Tipo_Eletronico'].fillna(moda_tipo)
df_padronizacao['Origem'] = df_padronizacao['Origem'].fillna(moda_origem)
df_padronizacao['Destino_Final'] = df_padronizacao['Destino_Final'].fillna(moda_destino)

# Conferência final
print("✅ Valores ausentes após imputação:")
print(df_padronizacao.isna().sum())

print("\n4️⃣ Verificação de duplicatas após padronização e imputação\n")

# Checamos se existem registros duplicados na base df_padronizacao
duplicatas_encontradas = df_padronizacao.duplicated().sum()

# Informamos o resultado
if duplicatas_encontradas > 0:
    print(f"⚠️ Foram encontradas {duplicatas_encontradas} duplicatas na base de dados.")
    print("As duplicatas podem ser analisadas com df_padronizacao[df_padronizacao.duplicated(keep=False)]")
else:
    print("✅ Nenhuma duplicata encontrada após a padronização e imputação de valores ausentes.")

# Conferir total de registros válidos restantes
print(f"📊 Total de registros válidos: {len(df_padronizacao)}")
```

✅ **Comentários sobre a Atividade 2 – Limpeza e Padronização:**

1. A coluna **Data_Descarte** foi convertida de `object` para `datetime`.  
   - Para facilitar visualização no padrão brasileiro, criamos a coluna **Data_Descarte_BR** no formato `dd/mm/yyyy`.  
   - Isso garante análises temporais consistentes.

2. Padronização de valores categóricos:  
   - **Tipo_Eletronico**: registros inconsistentes como `celular` e `Celular` foram unificados em `Celular`.  
   - **Origem**: valores como `Urbana` e `urbano` foram padronizados para `Urbano`.  
   - **Destino_Final**: `reciclagem formal` foi padronizado para `Reciclagem Formal`.  
   - **Municipio** e **Educacao_Ambiental** tiveram apenas limpeza de espaços extras.

3. Valores ausentes:  
   - Foram identificados **64 registros** com pelo menos um valor ausente, afetando principalmente as colunas `Tipo_Eletronico`, `Origem`, `Destino_Final`, `Peso_kg` e `Custo_Reciclagem_R$`.  
   - O preenchimento foi realizado com **mediana** para colunas numéricas e **moda** para colunas categóricas.

4. Duplicatas:  
   - Mesmo após a padronização, não foram identificados registros duplicados.

5. Qualidade geral da base:  
   - Colunas categóricas consistentes e uniformes.  
   - Colunas numéricas sem valores ausentes.  
   - Datas padronizadas e consistentes.  
   - Nenhum registro duplicado.

A base está agora pronta para análises estatísticas, relatórios e visualizações confiáveis.

# **:: Atividade 3 – Distribuição e Frequências**

<h4 align="justify">Antes de planejar políticas públicas de descarte e reciclagem, é essencial saber quais eletrônicos mais aparecem e de onde eles vêm.</h4>
<h4 align="justify">Utilize groupby e value_counts para identificar a frequência de cada tipo de eletrônico e de cada origem, apresentando os resultados em tabelas e em gráficos de barras (Matplotlib ou Seaborn).</h4>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Usando DataFrame tratado da Atividade 2 e criando outro para trabalharmos com os gráficos
df_sem_duplicata = df_padronizacao.copy()

# 1) Frequência por Tipo de Eletrônico (Matplotlib)
freq_tipo = df_sem_duplicata["Tipo_Eletronico"].value_counts()
print("\n=== 1️⃣ Frequência por Tipo de Eletrônico ===")
print(freq_tipo.to_frame("Frequência"))

plt.figure(figsize=(10,5))
plt.bar(freq_tipo.index, freq_tipo.values, color="steelblue", edgecolor="black")
plt.title("Frequência dos Tipos de Eletrônicos", fontsize=13)
plt.xlabel("Tipo de Eletrônico")
plt.ylabel("Quantidade")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 2) Frequência por Origem (Seaborn)
freq_origem = df_sem_duplicata["Origem"].value_counts()
print("\n=== 2️⃣ Frequência por Origem ===")
print(freq_origem.to_frame("Frequência"))

plt.figure(figsize=(8,5))
sns.barplot(x=freq_origem.index, y=freq_origem.values, palette="Greens_r")
plt.title("Frequência das Origens", fontsize=13)
plt.xlabel("Origem")
plt.ylabel("Quantidade")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 3) Cruzamento: Frequência de Tipo por Origem
# a) Tabela "long" com groupby
freq_por_origem = df_sem_duplicata.groupby(["Origem", "Tipo_Eletronico"]).size().reset_index(name="Frequência")
print("\n=== 3️⃣ Frequência por Origem e Tipo (tabela long) ===")
display(freq_por_origem.sort_values(["Origem","Frequência"], ascending=[True, False]).head(30))

# b) Pivot (Origem x Tipo)
pivot = freq_por_origem.pivot_table(index="Origem", columns="Tipo_Eletronico", values="Frequência", fill_value=0)
print("\n=== 4️⃣ Tabela pivot (Origem × Tipo_Eletronico) ===")
display(pivot)

# c) Gráfico agrupado (countplot) por origem
plt.figure(figsize=(12,6))
sns.countplot(df_sem_duplicata, x="Origem", hue="Tipo_Eletronico")
plt.title("Contagem de Tipos de Eletrônicos por Origem")
plt.xlabel("Origem")
plt.ylabel("Quantidade")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Tipo de Eletrônico", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

2. **Frequência por Origem:**

# **:: Atividade 5 – Agregação por Município**

<h4 align="justify">A gestão de resíduos eletrônicos exige que municípios compreendam seu papel na geração e no destino do lixo tecnológico.</h4>
<h4 align="justify">Utilize groupby para agregar os dados por município, calculando:</h4>
<h4 align="justify">&bullet; o total de resíduos descartados;</h4>
<h4 align="justify">&bullet; a média de peso dos eletrônicos; e</h4>
<h4 align="justify">&bullet; e a proporção de destinos formais (reciclagem, doação) em relação aos informais (lixão, descarte irregular).</h4>
<h4 align="justify">Apresente os resultados em uma tabela ordenada e em gráfico de barras horizontais.</h4>

```python
# 1. Criando cópia do dataset tratado para manipulação nesta atividade
df_atv5 = df_padronizacao.copy()

# 2. Agregar os dados por 'Municipio'
total_residuos = df_atv5.groupby('Municipio').size().rename('Total_Residuos')
media_peso = df_atv5.groupby('Municipio')['Peso_kg'].mean().rename('Media_Peso_kg')

# Todos os destinos finais: ['Aterro Controlado' 'Cooperativa' 'Reciclagem Formal' 'Lixão' 'Rio']
destinos_formais = ['Cooperativa', 'Reciclagem Formal']
df_atv5['Tipo_Destino'] = df_atv5['Destino_Final'].apply(lambda x: 'Formal' if x in destinos_formais else 'Informal')
contagem_destinos = df_atv5.groupby(['Municipio', 'Tipo_Destino']).size().unstack(fill_value=0)
proporcao_formal = (contagem_destinos.get('Formal', 0) / (contagem_destinos.get('Formal', 0) + contagem_destinos.get('Informal', 0))).rename('Proporcao_Formal')

# 3. Combinar e ordenar a tabela
tabela_agregada = pd.concat([total_residuos, media_peso, proporcao_formal], axis=1).fillna(0)
tabela_agregada = tabela_agregada.sort_values(by='Total_Residuos', ascending=False)

# CORREÇÃO DA TABELA: Formatar e imprimir
print("--- Tabela Agregada por Município ---")
tabela_formatada = tabela_agregada.copy()
tabela_formatada['Media_Peso_kg'] = tabela_formatada['Media_Peso_kg'].map('{:.2f}'.format)
tabela_formatada['Proporcao_Formal'] = tabela_formatada['Proporcao_Formal'].map('{:.2%}'.format)
print(tabela_formatada)

# 4. Apresentação em Gráficos de Barras Horizontais
fig, axes = plt.subplots(3, 1, figsize=(10, 15))
fig.suptitle('Análise Agregada por Município', fontsize=16)

# Gráfico 1: Total de Resíduos
sns.barplot(x=tabela_agregada['Total_Residuos'], y=tabela_agregada.index, hue=tabela_agregada.index, ax=axes[0], palette='viridis', legend=False)
axes[0].set_title('Total de Resíduos Descartados por Município')
axes[0].set_xlabel('Quantidade de Itens')
axes[0].set_ylabel('Município')

# Gráfico 2: Média de Peso
sns.barplot(x=tabela_agregada['Media_Peso_kg'], y=tabela_agregada.index, hue=tabela_agregada.index, ax=axes[1], palette='plasma', legend=False)
axes[1].set_title('Média de Peso (kg) dos Eletrônicos por Município')
axes[1].set_xlabel('Peso Médio (kg)')
axes[1].set_ylabel('Município')

# Gráfico 3: Proporção de Destinos Formais
sns.barplot(x=tabela_agregada['Proporcao_Formal'], y=tabela_agregada.index, hue=tabela_agregada.index, ax=axes[2], palette='magma', legend=False)
axes[2].set_title('Proporção de Descarte em Destinos Formais por Município')
axes[2].set_xlabel('Proporção (%)')
axes[2].set_ylabel('Município')
axes[2].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0%}'.format(x)))

plt.tight_layout()
plt.show()
```

```python
df_padronizacao_copia = df_padronizacao.copy() # para manutenção da integridade do que já foi tratado, vamos utilizar uma cópia

# 1. Criar a Tabela Cruzada (Crosstab)

tabela_cruzada = pd.crosstab(df_padronizacao_copia['Tipo_Eletronico'], df_padronizacao_copia['Destino_Final']) # quantidade em inteiro
tabela_cruzada.columns.name = None # Remover o nome da coluna Destino_Final

tabela_percentual = pd.crosstab(df_padronizacao_copia['Tipo_Eletronico'], df_padronizacao_copia['Destino_Final'], normalize='index') * 100 #traz a distribuição em percentual
tabela_percentual_formatada = tabela_percentual.style.format('{:.2f}%') # formatação para aparecer na tabela
tabela_percentual_formatada.columns.name = None # Remover o nome da coluna Destino_Final


print("--- Tabela Cruzada: Tipo de Eletrônico vs. Destino Final ---")
display(tabela_cruzada)

print("\n\n--- Tabela Cruzada: Tipo de Eletrônico vs. Destino Final (%) ---")
display(tabela_percentual_formatada)


# 2. Visualizar os resultados em um Heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(
    tabela_cruzada,
    annot=True,       # mostra os valores nas células
    cmap="YlOrRd",    # paleta de cores amarelo, laranja e vermelho, para destacarem pontos críticos
    fmt='g'           # formato dos números (g = inteiro)
)

print("\n--- Heatmap: Tipo de Eletrônico vs. Destino Final ---\n")
plt.title('Relação entre Tipo de Eletrônico e seu Destino Final', fontsize=24)
plt.xlabel('Destino Final', fontsize=12)
plt.ylabel('Tipo de Eletrônico', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

#também em percentual de eletrônico por destino final

plt.figure(figsize=(14, 8))
sns.heatmap(
    tabela_percentual,
    annot=True,       # mostra os valores nas células
    cmap="Oranges",    # paleta de tons alaranjados, para destacarem pontos críticos
    fmt='.2f'           # formato dos números (f = float)
)

print("\n--- Heatmap: Tipo de Eletrônico vs. Destino Final (%) ---\n")
plt.title('Relação entre Tipo de Eletrônico e seu Destino Final', fontsize=24)
plt.xlabel('Destino Final', fontsize=12)
plt.ylabel('Tipo de Eletrônico', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

# **:: Atividade 9 – Correlação entre Origem e Tipo de Eletrônico**

<h4 align="justify">A origem do resíduo eletrônico pode estar associada ao tipo de dispositivo descartado. Grandes centros urbanos tendem a gerar computadores, smartphones e notebooks, enquanto áreas rurais podem descartar rádios, televisores antigos e baterias automotivas. Essa correlação ajuda a entender os padrões de consumo e de descarte.</h4>
<h4 align="justify">Compreender tais relações é vital para desenhar políticas regionais: por exemplo, pontos de coleta de baterias em cidades pequenas ou centrais de reciclagem de computadores em polos universitários. Identificar as variáveis que caminham juntas pode revelar gargalos e oportunidades estratégicas.</h4>
<h4 align="justify">Calcule a correlação entre origem (urbana/rural/industrial) e tipo de eletrônico usando análise de contingência (qui-quadrado).</h4>
<h4 align="justify">Calcule a correlação entre origem (urbana/rural/industrial) e tipo de eletrônico usando análise de contingência (qui-quadrado).</h4>

```python
from scipy.stats import chi2_contingency

df_padronizacao_copia = df_padronizacao.copy() # para manutenção da integridade do que já foi tratado, vamos utilizar uma cópia

# 1. Criar a Tabela de Contingência (Crosstab)
tabela_contingencia = pd.crosstab(df_padronizacao_copia['Origem'], df_padronizacao_copia['Tipo_Eletronico'])

print("--- Tabela de Contingência: Origem vs. Tipo de Eletrónico ---")
print(tabela_contingencia)


# 2. Realizar o Teste Qui-Quadrado
chi2, p_value, dof, expected = chi2_contingency(tabela_contingencia)

print("\n--- Resultados do Teste Qui-Quadrado ---")
print(f"Estatística Qui-Quadrado (χ²): {chi2:.4f}")
print(f"Valor-p (p-value): {p_value:.4f}")
print(f"Graus de Liberdade (dof): {dof}")


# 3. Interpretação do Resultado
print("\n--- Interpretação ---")
alpha = 0.05  # Nível de significância comum

if p_value < alpha:
    print(f"Como o valor-p ({p_value:.4f}) é menor que {alpha}, rejeitamos a hipótese nula.")
    print("Conclusão: Existe uma associação estatisticamente significativa entre a Origem do resíduo e o Tipo de Eletrónico.")
else:
    print(f"Como o valor-p ({p_value:.4f}) não é menor que {alpha}, não podemos rejeitar a hipótese nula.")
    print("Conclusão: Não há evidências de uma associação estatisticamente significativa entre a Origem do resíduo e o Tipo de Eletrónico.")
# 1. Heatmap da tabela de contingência
plt.figure(figsize=(10,6))
sns.heatmap(tabela_contingencia, annot=True, cmap="Blues", fmt="d")
plt.title("Heatmap - Origem vs. Tipo de Eletrônico")
plt.ylabel("Origem")
plt.xlabel("Tipo de Eletrônico")
plt.show()

# 2. Gráfico de barras agrupadas
tabela_contingencia.plot(kind="bar", figsize=(10,6))
plt.title("Distribuição dos Tipos de Eletrônicos por Origem")
plt.ylabel("Frequência")
plt.xlabel("Origem")
plt.xticks(rotation=0)
plt.legend(title="Tipo de Eletrônico")
plt.show()
```

*   O descarte de eletrônicos parece relativamente homogêneo nas diferentes origens.

*   Por exemplo, tanto áreas Governamentais quanto Urbanas apresentam descartes de baterias e celulares, enquanto o setor Industrial e Rural também descarta computadores e TVs em proporções parecidas.

*   Não existe evidência forte de que, por exemplo, "o meio rural descarta mais tablets que o urbano" ou "o governo descarta mais geladeiras que as indústrias".

### Chegando a conclusao que :
Não há indícios de que a origem do resíduo eletrônico esteja associada ao tipo de eletrônico descartado.
Isso sugere que o problema do descarte de eletrônicos é generalizado e não restrito a um único setor (todos descartam de forma parecida).

### Insight 1 – Operação Sincronizada de Captura de Valor
Existe uma janela anual (dezembro-janeiro) onde grandes volumes de resíduos de alto valor são descartados de forma inadequada.

- **Interceptação Temporal:** agendamento obrigatório de descartes em dezembro-janeiro, rotas sincronizadas e triagem temporária.  
- **Monetização do Peso-Valor:** bonificação por peso, cooperativas especializadas e créditos de reciclagem.  
- **Aproveitamento da Infraestrutura:** transporte fluvial reverso, coleta em prédios públicos e logística reversa com fornecedores.

**Impacto Esperado:** captura de 60% dos resíduos de maior valor em 60 dias, criando modelo replicável.

### Insight 2 – Dupla Falha Logística
O descarte governamental massivo e previsível e o descarte urbano difuso configuram uma dupla falha que exige solução integrada.

Essa dispersão evidencia a necessidade de padronização e políticas específicas por origem, reforçando a importância de programas integrados como Amazônia Circular.

**Proposta Prática: Programa Amazônia Circular**
- **Frente institucional:** coleta agendada via plataforma online, realizada por cooperativas certificadas.  
- **Frente urbana – Recicla Bairro:** pontos de recolha inteligentes (PRI), agendamento domiciliar e incentivos gamificados (ex.: descontos em contas).  
- **Logística integrada:** mapeamento de rotas, incluindo transporte fluvial, para conectar pontos de coleta a centros de triagem.
